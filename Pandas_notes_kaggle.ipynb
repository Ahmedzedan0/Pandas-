```python
specific_house_data = house_data.mean()
specific_house_data.map(lambda p: p - specific_house_data)
specific_house_data 
```

```python
def remean_points(row):
    row.Distance = row.Distance - specific_house_data
    return row

house_data.apply(remean_points, axis='columns')
```
